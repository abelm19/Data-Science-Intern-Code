**Exploring Drowsiness Patterns in Wearable Device Data **

---

* heartRate: Heart rate readings from the smartwatch sensors.
* ppgGreen, ppgRed, ppgIR: PPG (Photoplethysmography) sensor readings in green,
red, and infrared wavelengths respectively.
* drowsiness: Label indicating the level of drowsiness based on an adapted Karolinska
Sleepiness Scale (KSS). Values range from 0.0 to 2.0, where 0.0 represents
alertness and 2.0 represents significant drowsiness.

Period Marked Analysis using Scatter Plots, Periods Include Morning, Afternoon, Evening & Night

```python
#Dividing the data into 4 periods labeled as Morning, Afternoon, Evening and Night equal datasubsets
num_periods = 4
period_size = len(data) // num_periods
period_labels = ['Morning', 'Afternoon', 'Evening', 'Night']
data['period'] = pd.cut(data.index, bins = num_periods, labels = period_labels)
print(data['period'].value_counts())

morning_data = data[data['period'] == 'Morning']
afternoon_data = data[data['period'] == 'Afternoon']
evening_data = data[data['period'] == 'Evening']
night_data = data[data['period'] == 'Night']

#calculates the correlation for each period
#and creates a plot also for each period
def calculate_and_plot_cor(data, period):

    corrdrows_heartRate = data['drowsiness'].corr(data['heartRate'])
    corrdrows_ppgGreen = data['drowsiness'].corr(data['ppgGreen'])
    corrdrows_ppgRed = data['drowsiness'].corr(data['ppgRed'])
    corrdrows_ppgIR = data['drowsiness'].corr(data['ppgIR'])

    print(f'Correlation between drowsiness and heartRate for ({period}) time is ({corrdrows_heartRate})')
    print(f'Correlation between drowsiness and ppgGreen for ({period}) time is ({corrdrows_ppgGreen})')
    print(f'Correlation between drowsiness and ppgRed for ({period}) time is ({corrdrows_ppgRed})')
    print(f'Correlation between drowsiness and ppgIR for ({period}) time is ({corrdrows_ppgIR})')

    plt.scatter(data['ppgGreen'], data['drowsiness'], alpha = 0.5, label = 'PPG Green', color = 'green')
    plt.scatter(data['ppgRed'], data['drowsiness'], alpha = 0.5, label = 'PPG Red', color = 'red')
    plt.scatter(data['ppgIR'], data['drowsiness'], alpha = 0.5, label = 'PPG IR', color = 'blue')
    plt.title(f'Drowsiness Levels and PPG Readings ({period})')
    plt.xlabel('Activity Level')
    plt.ylabel('Drowsiness Level')
    plt.legend(loc = 'lower left')
    plt.show()

calculate_and_plot_cor(morning_data, "Morning")
calculate_and_plot_cor(afternoon_data, 'Afternoon')
calculate_and_plot_cor(evening_data, 'Evening')
calculate_and_plot_cor(night_data, 'Night')


#We confirm that almost in all the periods Morning to Night, heartRate and drowsiness have negative correlation values
#meaning that as heartRate decreases, Drowsiness is likely to increase, so the smartwatch can alert the user
#by reading the heartRate accurately we can notify the user of potential drowsiness

#But as for the PPG indicators, PPG Green seems to have no to little correlation with drowsiness, indicating that
# maybe it might not be a good indicator to alert the user of drowsiness also having the most outliers.
# In the case of PPG Red it seems to have a negative correlation in Morning and Evening time and not so much
# a correlation during Afternoon and Night times
# Finally, in the case of PPG IR it seems also same with PPG Red, it seems to have a negative correlation in Morning and Evening time and not so much
# a correlation during Afternoon and Night times
```
